## Exercises ! 

```python
#Write a loop that counts the number of vowels in a character string.

vowels = "aeiouy"

my_string = "hello, i'm Ida AAAEEHH"

vowel_count = 0

for vowel in my_string.lower(): 
    if vowel in vowels:
        vowel_count += 1

print(f'The number of vowels in this string is {vowel_count}')
```
